# Problem 1

```python
#Q1. Single‐server queue system according to a nonhomogeneous Poisson process.

import numpy as np

time_arr = []
lst = []
est = 0
break_arr = []
rate = [x for x in range(4, 20, 3)] + [x for x in range(19, 6, -3)]
for i in range(100):
    count = i
    while True:
        count += (-1/rate[i % 10]) * np.log(np.random.uniform())
        if count - i > 1:
            break
        time_arr.append(count) 

while len(time_arr):
    if est < time_arr[0]:
        break_arr.append(0.3 * np.random.rand())
        est += break_arr[-1]
    else:
        est += np.random.exponential(1 / 25, 1)
        time_arr.remove(time_arr[0])
lst.append(sum(break_arr))
print("The expected amount of time that the server is on break =", np.mean(lst))
```

# Problem 2

```python
#Q2. HOL‐blocking switch

import numpy as np
import matplotlib.pyplot as plt

buf_1 = []
buf_2 = []
cond1 = []
cond2 = []
r = [0.75, 0.25]
for p in np.arange(0.01,1,0.01):
    buffer1 = 0
    buffer2 = 0
    output1 = 0
    output2 = 0
    for i in range(1000):
        if buffer1 and buffer2:
            choose1 = np.random.uniform()
            choose2 = np.random.uniform()

            if choose1 < r[0] and choose2 < r[0]:
                output1 += 1
                if np.random.uniform() < 0.5:
                    buffer1 -= 1
                else:
                    buffer2 -= 1

            elif choose1 >= r[0] and choose2 >= r[0]:
                output2 += 1
                if np.random.uniform() < 0.5:
                    buffer1 -= 1
                else:
                    buffer2 -= 1

            else:
                output1 += 1
                output2 += 1
                buffer1 -= 1
                buffer2 -= 1

        elif buffer1:
            buffer1 -= 1
            if np.random.uniform() < r[0]:
                output1 += 1
            else:
                output2 += 1

        elif buffer2:
            buffer2 -= 1
            if np.random.uniform() < r[0]:
                output1 += 1
            else:
                output2 += 1

        if np.random.uniform() < p:
            buffer1 += 1
        if np.random.uniform() < p:
            buffer2 += 1
    buf_1.append(buffer1)
    buf_2.append(buffer2)
    cond1.append(output1/1000)
    cond2.append(output2/1000)

plt.figure()
plt.plot(np.arange(0.01,1,0.01), cond1)
plt.xlabel('arrival probability p')
plt.ylabel('mean packets processed')

plt.figure()
plt.plot(np.arange(0.01, 1, 0.01), cond2)
plt.xlabel('arrival probability p')
plt.ylabel('mean packets processed')
plt.show()
```
