# Random number generator for mixture distribution

```python
#Random number generator for mixture distribution

import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

N = 1000
p = 0.6
x = np.linspace(-5, 5, 100)
y = 0.4 * np.array(norm(-1, 1).pdf(x)) + 0.6 * np.array(norm(1, 1).pdf(x))
s = []
for i in range(N):
    if np.random.rand() < p:
        s.append(np.random.normal(1,1))
    else:
        s.append(np.random.normal(-1,1))
plt.figure()
plt.hist(s, bins=20, density=True, edgecolor='black', label='simulation')
plt.plot(x, y.tolist(), label='theoretical', color='r')
plt.title("histogram and theoretical pdf of mixture")
plt.legend()
plt.show()
```
