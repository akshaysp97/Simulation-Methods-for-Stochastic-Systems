# To find the 25th, 50th, and 90th percentiles using empirical distribution and compare the value to the theoretical percentile values for ùúí2(4)

```python
#Emperical CDF

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


for n in [10, 100, 1000]:
    result = []
    for i in range(n):
        z1 = np.random.standard_normal()
        z2 = np.random.standard_normal()
        z3 = np.random.standard_normal()
        z4 = np.random.standard_normal()
        result.append(np.square(z1)+np.square(z2)+np.square(z3)+np.square(z4))
    
    x = np.sort(result)
    y = np.arange(0,1,1/n)
    lst = np.arange(min(x),max(x),0.01)
    cdf = stats.chi2.cdf(lst, df=4)
    plt.figure()
    plt.title('n = ' + str(n) + ' Empirical and theoretical CDF')
    plt.step(x, y, label = 'Empirical CDF')
    plt.plot(lst, cdf, label ='Theoretical CDF', color='red')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.show()
    
    diff = max(abs(stats.chi2.cdf(x,4)-np.arange(1/n,1+1/n,1/n)))
    print('n =', n)
    print('lower bound =', diff)
    print('Theoretical 25th percentile =', stats.chi2.ppf(0.25, df=4))
    print('Theoretical 50th percentile =', stats.chi2.ppf(0.5, df=4))
    print('Theoretical 90th percentile =', stats.chi2.ppf(0.9, df=4))
    print('\nEmpirical 25th percentile =', np.percentile(x[int(n/4)],25))
    print('Empirical 50th percentile =', np.percentile(x[int(n/2)], 50))
    print('Empirical 90th percentile =', np.percentile(x[int(9*n/10)], 90))
    print('\n')
```
