# Sampling alpha-stable random variable

```python
#Sampling alpha-stable random variable

import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import levy_stable

def inv(w, a, b):
    gamma = np.random.uniform(-np.pi/2, np.pi/2)
    if a == 1:
        return (np.pi/2 + b*gamma)*math.tan(gamma)-b*np.log(w*math.cos(gamma)/(np.pi/2 + b*gamma))
    else:
        gamma_0 = (-np.pi/2)*b*dec(a)/a
        return (math.sin(a*(gamma-gamma_0))/(math.cos(gamma)**(1/a))) * ((math.cos(gamma-a*(gamma-gamma_0))/w)**((1-a)/a))
    
def draw(a, b, fig):
    lst = []
    for i in range(1000):
        w = np.random.exponential(1)
        lst.append(inv(w, a, b))
        x = np.arange(-10,10,0.1)
    plt.figure(fig)
    plt.subplot(1,2,1)
    plt.hist(lst, bins=20, range=(-10,10),edgecolor="black", color='r', label='a='+str(a)+' b='+str(b),rwidth=0.8, density=1)
    plt.plot(x,levy_stable.pdf(x,a,b),label="Theoretical",color="b")
    plt.title('Sample histogram')
    plt.legend()
    plt.subplot(1,2,2)
    plt.plot(np.arange(0,1000), lst)
    plt.title('time series plot')
    return

def dec(a):
    if a < 1:
        return a
    elif a > 1:
        return a-2

alpha = [0.5, 1, 1.8, 2.0]
beta = [0, 0.75]
fig = 1
for i in beta:
    for j in alpha:
        draw(j, i, fig)
        fig += 1
plt.show()
```
