# Sampling gamma random variable

```python
#Sampling gamma random variable

import numpy as np
import math
import matplotlib.pyplot as plt

def f(x, a, b):
    return (x ** (a - 1)) * ((1 / b) ** a) * (np.exp(-x / b)) / ((math.pi ** 0.5) * 1 * 3 * 5 * 7 * 9/(2**5))

def g(x):
    return (2/11)*np.exp(-2*x/11)

N = 1000
lst = []
count = 0
for i in range(N):
    u1 = np.random.uniform()
    u1 = -5.5 * np.log(u1)
    if np.random.uniform() < f(u1, 5.5, 1) / (np.e * g(u1)):
        count += 1
        lst.append(u1)
pdf = []
x = np.arange(0, 20, 0.5)
for i in x:
    pdf.append(f(i, 5.5, 1))

plt.figure()
plt.hist(lst, bins=20, edgecolor='black', label='simulation', density=1)
plt.plot(x, pdf, color='r', label='theoretical')
plt.legend()
plt.show()
print('acceptance rate =', count/N)
```
